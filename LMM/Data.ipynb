```python
import pandas as pd
import numpy as np
import pandas as pd
import os
from matplotlib import image
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
hv = pd.read_csv('/Users/reesaespera/Desktop/tjl/Havard Dataset/Full_Descriptors.csv')
nf = pd.read_csv('/Users/reesaespera/Desktop/tjl/Non-Fullerene Dataset/NF_Descriptors.csv')
cp = pd.read_csv('/Users/reesaespera/Desktop/tjl/Conjugated Polymers Dataset/CP_Data.csv')
dd = pd.read_csv('/Users/reesaespera/Desktop/tjl/Dopant Dataset/Dopant_Data.csv')
```

```python
hv['Group'] = '1'
nf = nf.rename(columns={'smiles':'SMILES', 'HOMO_calc':'HOMO', 'LUMO_calc':'LUMO'}).drop(columns=['LUMO_calib','LUMO_calib_stds','HOMO_calib','HOMO_calib_stds'])
nf['Group'] = '2'
cp = cp.rename(columns={'-HOMO(eV)':'HOMO','-LUMO(eV)':'LUMO', 'smiles':'SMILES'}).drop(columns=['IDNo.', 'Nickname', 'Ref.No', 'PCE_max(%)', 'PCE_ave(%)','Voc(V)','Jsc(mAcm^2)','FF','Mw(kgmol^-1)','Mn(kgmol^-1)','PDI(=Mw/Mn)','Monomer(gmol^-1)','bandgap(eV)','mol'])
cp['Group'] = '3'
dd = dd.rename(columns={'smiles':'SMILES'}).drop(columns=['Unnamed: 0','HOMO - 1','LUMO + 1'])
dd['Group'] = '4'
```

```python
combo = pd.concat([hv, dd, cp, nf], ignore_index=True)
columns = ['HOMO', 'LUMO', 'SMILES', 'Molecular Weight', 'Heavy Atom Molecular Weight', 'Max Absolute Partial Charge', 'Max Partial Charge', 
         'Min Abs Partial Charge', 'Min Partial Charge', 'Radical Electrons', 'Valence Electrons', 'NHOH Count', 'NO Count',
         'H Acceptors', 'H Donors', 'Ring Count', 'Aliphatic Rings', 'Aromatic Rings', 'Saturated Rings', 'Aromatic Carbocycles',
         'Aromatic Heterocycles', 'Heteroatoms', 'Rotatable Bonds', 'Saturated Carbocycles', 'Saturated Heterocycles', 'H Count',
         'C Count', 'N Count', 'F Count', 'Halogen Count', 'Double Bonds', 'Triple Bonds']

combo = combo.rename(columns={'Molecular Weight':'MolWt', 'Heavy Atom Molecular Weight':'HvAtMolWt', 
                            'Max Absolute Partial Charge':'MaxAbsPartCh', 'Max Partial Charge':'MaxPartCh',
                            'Min Abs Partial Charge':'MinAbsPartCh', 'Radical Electrons':'RadE', 'Valence Electrons':'ValE',
                            'NHOH Count':'NHOH', 'NO Count':'NO', 'H Acceptors':'HAcc', 'H Donors':'HDon', 'Ring Count':'Rings',
                            'Aliphatic Rings':'AlRing', 'Aromatic Rings':'AroRing', 'Saturated Rings':'SatRing',
                            'Aromatic Carbocycles':'AroCarb', 'Aromatic Heterocycles':'AroHet', 'Rotatable Bonds':'RotBond',
                            'Saturated Carbocycles':'SatCarb', 'Saturated Heterocycles':'SatHet', 'H Count':'H', 'C Count':'C',
                            'N Count':'N', 'F Count':'F', 'Halogen Count':'Halogens', 'Double Bonds':'Double', 'Triple Bonds':'Triple',
                            'Min Partial Charge':'MinPartCh'})
# combo = combo.dropna(axis=1)

combo = combo.drop(columns=['Unnamed: 0', 'Unnamed: 0.1', 'mol'])
```
