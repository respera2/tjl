```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Lasso
import matplotlib.pyplot as plt
import matplotlib
import rdkit.Chem.Descriptors as d
from rdkit import Chem
```

```python
f = pd.read_csv('Dopants.csv')

f['mol'] = f['SMILES String'].apply(lambda x: Chem.MolFromSmiles(x))
f = f.dropna()
print(f)

f['smiles'] = f['mol'].apply(lambda x: Chem.MolToSmiles(x))
f = f.drop(['SMILES String'], axis=1)
f = f.drop_duplicates(subset='smiles')


f['Molecular Weight'] = f['smiles'].apply(lambda x: d.ExactMolWt(Chem.MolFromSmiles(x)))
f['Heavy Atom Molecular Weight'] = f['smiles'].apply(lambda x: d.HeavyAtomMolWt(Chem.MolFromSmiles(x)))
f["Max Absolute Partial Charge"] = f['smiles'].apply(lambda x: d.MaxAbsPartialCharge(Chem.MolFromSmiles(x)))
f["Max Partial Charge"] = f['smiles'].apply(lambda x: d.MaxPartialCharge(Chem.MolFromSmiles(x)))
f["Min Abs Partial Charge"] = f['smiles'].apply(lambda x: d.MinAbsPartialCharge(Chem.MolFromSmiles(x)))
f["Min Partial Charge"] = f['smiles'].apply(lambda x: d.MinPartialCharge(Chem.MolFromSmiles(x)))
f["Molecular Weight"] = f['smiles'].apply(lambda x: d.MolWt(Chem.MolFromSmiles(x)))
f["Radical Electrons"] = f['smiles'].apply(lambda x: d.NumRadicalElectrons(Chem.MolFromSmiles(x)))
f["Valence Electrons"] = f['smiles'].apply(lambda x: d.NumValenceElectrons(Chem.MolFromSmiles(x)))
f["NHOH Count"] = f['smiles'].apply(lambda x: d.NHOHCount(Chem.MolFromSmiles(x)))
f["NO Count"] = f['smiles'].apply(lambda x: d.NOCount(Chem.MolFromSmiles(x)))
f["H Acceptors"] = f['smiles'].apply(lambda x: d.NumHAcceptors(Chem.MolFromSmiles(x)))
f["H Donors"] = f['smiles'].apply(lambda x: d.NumHDonors(Chem.MolFromSmiles(x)))
f["Ring Count"] = f['smiles'].apply(lambda x: d.RingCount(Chem.MolFromSmiles(x)))
f["Aliphatic Rings"] = f['smiles'].apply(lambda x: d.NumAliphaticRings(Chem.MolFromSmiles(x)))
f["Aromatic Rings"] = f['smiles'].apply(lambda x: d.NumAromaticRings(Chem.MolFromSmiles(x)))
f["Saturated Rings"] = f['smiles'].apply(lambda x: d.NumSaturatedRings(Chem.MolFromSmiles(x)))
f["Aromatic Carbocycles"] = f['smiles'].apply(lambda x: d.NumAromaticCarbocycles(Chem.MolFromSmiles(x)))
f["Aromatic Heterocycles"] = f['smiles'].apply(lambda x: d.NumAromaticHeterocycles(Chem.MolFromSmiles(x)))
f["Heteroatoms"] = f['smiles'].apply(lambda x: d.NumHeteroatoms(Chem.MolFromSmiles(x)))
f["Rotatable Bonds"] = f['smiles'].apply(lambda x: d.NumRotatableBonds(Chem.MolFromSmiles(x)))
f["Saturated Carbocycles"] = f['smiles'].apply(lambda x: d.NumSaturatedCarbocycles(Chem.MolFromSmiles(x)))
f["Saturated Heterocycles"] = f['smiles'].apply(lambda x: d.NumSaturatedHeterocycles(Chem.MolFromSmiles(x)))
```

```python
from mordred import AtomCount
from mordred import BondCount

f["H Count"] = f['smiles'].apply(lambda x: AtomCount.AtomCount('H')(Chem.MolFromSmiles(x)))
f["C Count"] = f['smiles'].apply(lambda x: AtomCount.AtomCount('C')(Chem.MolFromSmiles(x)))
f["N Count"] = f['smiles'].apply(lambda x: AtomCount.AtomCount('N')(Chem.MolFromSmiles(x)))
f["F Count"] = f['smiles'].apply(lambda x: AtomCount.AtomCount('F')(Chem.MolFromSmiles(x)))
f["Halogen Count"] = f['smiles'].apply(lambda x: AtomCount.AtomCount('X')(Chem.MolFromSmiles(x)))
f["Double Bonds"] = f['smiles'].apply(lambda x: BondCount.BondCount('double', False)(Chem.MolFromSmiles(x)))
f["Triple Bonds"] = f['smiles'].apply(lambda x: BondCount.BondCount('triple', False)(Chem.MolFromSmiles(x)))
```

```python
f.to_csv('Dopant_Data.csv')
```
